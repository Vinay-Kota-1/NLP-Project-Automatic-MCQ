```python
from torch._C import NoneType
def disambiguity(sentence, keyword):
  '''
  This function creates a disambiguity scores and generates ambigous pairs for different keywords
  '''

  keyword = keyword.lower()
  sentence =sentence

  keyword = keyword.replace(" ","_")

  synsets_noun = wordnet.synsets(keyword,'n')
  synsets_verbs = wordnet.synsets(keyword,'v')

  synsets_comb = synsets_noun.extend(synsets_verbs)

  max = 0

  simalar_synset = 0

  if synsets_comb is None : return None
  else: 

    for i in synsets_comb:
     for word in sent_tokenize(sentence):
        sim = i.wup_similarity(word)
        if sim > max:
          max = sim
          simalar_synset = i

  return simalar_synset
```

```python
def distractors(synset, keyword):
  '''
  This function generates the synsets, and the corresponding 
  distractors. It returns the list of distractors for a corresponding keywords.
  This function takes context wise pairs.
  '''
  options = []
  a =  keyword
  keyword = keyword.lower()

  keyword = keyword.replace(" ","_")

  if synset is None: return None

  hypernyms_list = synset.hypernyms()

  if hypernyms_list[0] is None: return None
  else:

   

    for i in hypernyms_list.hyponyms()[0]:
    
      word = i.lemmas()[0].name()

      if word != a: options.append()

      else: continue

  return options
```

```python
def generate_questions(summary, dictionary, just_words):
  '''
  This one takes the summary, dictionary, keywords and
  generate sentences necessary for the question to be produced
  '''

  options = {}

  for k in dictionary:
    ambiguity = disambiguity(dictionary[k][0],k)

    distractor = distractors(ambiguity, k)

    if len(distractor) > 0:
      options[k] = distractor

    for i in options:
      obtained_sentence = dictionary[i][0]

      output_sentence = obtained_sentence.replace("i",'______')

      selections = [i] + options[i]

      output_4 = selections[:4]

      print(">>",output_sentence)

      for i,j  in enumerate(output_4):
        print(f'{i}: {j}')













```

```python
class BertForQuestionGeneration(nn.Module):
  def __init__(self, token_df, model, tokenizer):
    super(BertForQuestionGeneration, self).__init__()
    self.model = model
    self.tokenizer = tokenizer
    self.length = len(token_df)

    self.model_extend = nn.Sequential(
        nn.Flatten()
        , nn.Linear(512, 12),
        nn.Softmax(dim=1)
    )

  def forward(self, Input_tokens, input_attentions, output_encoded,output_attentions ):
    output = model(Input_tokens,input_attentions,output_encoded,output_attentions)

    outputs= self.model_extend(output)

    return outputs


```
